```python
import pandas as pd
import numpy as np
```

```python
files = ['MOCK_DATA (1).csv','MOCK_DATA (2).csv',
        'MOCK_DATA (3).csv','MOCK_DATA (4).csv','MOCK_DATA (5).csv',
         'MOCK_DATA (6).csv','MOCK_DATA (7).csv','MOCK_DATA (8).csv','MOCK_DATA (9).csv']
```

```python
df = pd.read_csv('MOCK_DATA.csv')

for file in files:
    df1 = pd.read_csv(file)
    df = pd.concat([df,df1],ignore_index=True)
```

```python
df
```

```python
time = pd.to_timedelta(np.random.randint(0,60*60*22,size=len(df)),unit='s')
```

```python
df['order_time']=pd.to_datetime(df['order_time']).add(time)
```

```python
del_time = pd.to_timedelta(np.random.randint(0,60*60,size = len(df)),unit='s')
df['delivery_time']=df['order_time'].add(del_time)
```
